```python
import pandas as pd
import numpy as np
import codecs
```

## データ読み込み

```python
with codecs.open("../../data/input/JPvideos.csv", "r", "utf-8", "ignore") as file:
    df = pd.read_table(file, delimiter=",")
channel_id_no_data = pd.read_csv("../../data/input/channel_id_no_data.csv")
subscriber = pd.read_csv("../../data/input/subscriber.csv")
```

## チャンネルIDのデータを作成

```python
channel_id = channel_id_no_data[channel_id_no_data["channel_id"]!="No Data"]
channel_id = channel_id.drop_duplicates()
channel_id = channel_id[["video_id", "channel_id"]]
```

```python
len(channel_id["video_id"].unique())
```

```python
len(channel_id["channel_id"].unique())
```

## チャンネル登録者数のデータを見てみる

```python
subscriber = subscriber[["channel_id", "subscriber_count"]]
```

```python
subscriber
```

## データを結合させる

```python
subscriber_and_channel_id = pd.merge(channel_id, subscriber, on="channel_id", how="inner").drop_duplicates()
```

```python
subscriber_and_channel_id
```

```python
df = pd.merge(df, subscriber_and_channel_id, on="video_id").drop_duplicates()
```

```python
df = df[df["subscriber_count"] != "No Data"]
```

```python
df = df.astype({"subscriber_count": float})
```

```python
df.info()
```

```python
df.to_csv("../../data/input/useVideos.csv")
```
